## define the feature functions

- feature vector is defined by $(1, f_1(x, y), f_2(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_feature(point):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    ones =np.ones(len(point))
    feature = np.c_[ones,point,point[:,0]*point[:,1],point**2,(point[:,0]**2)*(point[:,1]**2),point[:,0]*(point[:,1]**2),point[:,1]*(point[:,0]**2),(point[:,0]**2)*(point[:,1]**3),(point[:,1]**2)*(point[:,0]**3),point**3,(point[:,0]**3)*(point[:,1]**3)]
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return feature
```

## define the linear regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature = $(1, f_1(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_linear_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = np.inner(theta,feature)
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return value
```

## define sigmoid function with input

- $z \in \mathbb{R}$

```python
def sigmoid(z):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = 1 / (1 + np.exp(-z))
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value 
```

## define the logistic regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$

```python
def compute_logistic_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = sigmoid(compute_linear_regression(theta,feature))
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value
```

## define the residual function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_residual(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    h = compute_logistic_regression(theta, feature)
    residual = -label*np.log(h)-(1-label)*np.log(1-h)
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return residual
```

## define the loss function for the logistic regression

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_loss(theta, feature, label, alpha):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    residual=compute_residual(theta, feature, label)
    loss= residual.mean()+alpha/2*np.inner(theta.T,theta)
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return loss
```

## define the gradient of the loss with respect to the model parameter $\theta$

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_gradient(theta, feature, label, alpha):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    h=compute_logistic_regression(theta, feature)
    gradient =  np.dot((h - label).T, feature) / len(feature) +alpha*theta

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return gradient
```

## compute the accuracy of the prediction for point with a given model parameter

```python
def compute_accuracy(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    predicted_label = compute_logistic_regression(theta, feature)>=0.5
    diff = predicted_label- label
    accuracy = 1.0-np.count_nonzero(diff)/len(feature)  
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return accuracy
```

## initialize the gradient descent algorithm

```python
number_iteration    = 1000 # you can change this value as you want 
learning_rate       = 0.1 # you can change this value as you want 
number_feature      = 14 # you can change this value as you want
alpha               = 1 # you can change this value as you want

theta                       = np.zeros(number_feature)
loss_iteration_train        = np.zeros(number_iteration)
loss_iteration_test         = np.zeros(number_iteration)
accuracy_iteration_train    = np.zeros(number_iteration)
accuracy_iteration_test     = np.zeros(number_iteration)
```

## run the gradient descent algorithm to optimize the loss function with respect to the model parameter

```python
for i in range(number_iteration):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    feature = compute_feature(data_train_point)
    theta           = theta-learning_rate*compute_gradient(theta,feature,data_train_label,alpha)
    loss_train      = compute_loss(theta, feature, data_train_label, alpha)
    loss_test       = compute_loss(theta, feature, data_test_label, alpha)
    accuracy_train  = compute_accuracy(theta, feature, data_train_label)
    accuracy_test   = compute_accuracy(theta, feature, data_test_label)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    loss_iteration_train[i]     = loss_train
    loss_iteration_test[i]      = loss_test
    accuracy_iteration_train[i] = accuracy_train
    accuracy_iteration_test[i]  = accuracy_test

theta_optimal = theta
```
